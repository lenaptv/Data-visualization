# Матрица рассеяния (scatter matrix)

Рассмотрим пример ИТ-школы.

У студентов есть доступ к статистике по коммитам однокурсников, данная статистика ведется в ленте новостей.
Ранее было проверено, что студенты, которые имели доступ к данной ленте новостей (1ая группа студентов) раньше начинали выполнять задание, делать первые коммиты значительно заранее до дедлайна, чем те студенты, которые не имели доступа к ленте новостей (2ая группа студентов).
Поэтому доступ к ленте новостей был предоставлен всем студентам.

Но теперь организаторам школы важно понять, есть ли взаимосвязь не только между самим фактом просмотра ленты и размером средней дельты (средней по разным лабораторным работам) между дедлайном и первым коммитом (в данном примере, эта гипотеза уже была подтверждена), но и есть ли зависимость между количеством просмотров ленты и количеством коммитов, а также величиной дельты. При этом важно отметить, что чем больше дельта в абсолютной величине между дедлайном по лабораторной работе и первым коммитом, тем лучше, тк это означает, что у студента было больше времени разобраться в задании и попробовать разные подходы к решению.

Понять, есть ли зависимость между количеством просмотров ленты, размером дельты и количеством коммитов, поможет матрица рассеяния. Тем более, если эта зависимость нелинейная, тогда ее не покажет и коэффициент корреляции. 


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# по каждому студенту есть статистика по количеству коммитов, количеству просмотров ленты новостей 
# и размеру дельты (чем больше размер дельты в абсолютной величине, тем лучше)
df= pd.read_csv('data/avg_diff_and_views.csv')
df.head(2)
```

```python
df.shape
```

```python
# рисуем красивый график

pd.plotting.scatter_matrix(df, figsize=(15,8), diagonal='kde', s=200, color='r')
plt.rc('lines', linewidth=3) # задаем ширину линий
plt.show()
```

### Выводы

- существует много студентов с небольшим количеством коммитов и мало - с большим количеством коммитов,
- взаимосвязи, что чем больше у студента просмотров статистики, то тем больше коммитов - не выявлено,
- взаимосвязи, если у пользователя меньше просмотров страницы, то, скорее всего, у него будет небольшое среднее значение дельты (разница между датой первого коммита и сроком сдачи лабораторного задания) - также не выявлено.
